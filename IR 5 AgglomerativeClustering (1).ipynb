```python
#step 1 -- import libraries 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage

#step 2 -- generate data using make_blobs
X,y=make_blobs(n_samples=100,centers=3,cluster_std=1.05,random_state=42)

# step3 -- 3. Visualize Data
plt.scatter(X[:,0],X[:,1],c='blue',marker='o',edgecolor='black',s=50)
plt.title("Generated Data")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.show()

# Step 4 --> Apply Agglomerative Cust
agglomarative_clustering=AgglomerativeClustering(n_clusters=3, metric='euclidean',linkage='ward')
labels=agglomarative_clustering.fit_predict(X)

# Step 5 --> visualize Clustering result
# Plot clustering results
plt.scatter(X[:, 0], X[:, 1], c=labels, marker='o', edgecolor='black', s=50, cmap='viridis')
plt.title("Agglomerative Clustering Results")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
